```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('stock.csv')

# Histogram for Purchase Price
plt.figure(figsize=(8, 5))
plt.hist(data['Purchase Price'], bins=10, edgecolor='black')
plt.title('Distribution of Purchase Price')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.show()

# Histogram for Current Price
plt.figure(figsize=(8, 5))
plt.hist(data['Current Price'], bins=10, edgecolor='black')
plt.title('Distribution of Current Price')
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.show()

# Histogram for Gain/Loss
plt.figure(figsize=(8, 5))
plt.hist(data['Gain/Loss'], bins=10, edgecolor='black')
plt.title('Distribution of Gain/Loss')
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.show()
```

```python
#Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi
top_dividends = data.nlargest(5, 'Dividend/share')

top_stock_names = top_dividends['Stock Name'].values
top_symbols = top_dividends['Symbol'].values

print("Stock Name dan Symbol yang memiliki Dividend/Share tertinggi:")
for i in range(len(top_stock_names)):
    print(f"Stock Name: {top_stock_names[i]}, Symbol: {top_symbols[i]}, Dividend/Share: {top_dividends['Dividend/share'].values[i]}")
```

```python
#Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi
top_market_values = data.nlargest(5, 'Market Value')

top_stock_names = top_market_values['Stock Name'].values
top_symbols = top_market_values['Symbol'].values

print("Stock Name and Symbol yang memiliki Market Value tertinggi:")
for i in range(len(top_stock_names)):
    print(f"Stock Name: {top_stock_names[i]}, Symbol: {top_symbols[i]}, Market Value: {top_market_values['Market Value'].values[i]}")
```
